### Tecnologías de Procesamiento Big Data

```python
from TradingviewData import TradingViewData, Interval

request = TradingViewData()
```

```python
eth_data = request.get_hist(
    symbol='ETHUSDT',
    exchange='BINANCE',
    interval=Interval.daily,
    n_bars=1460
)
for year, group in eth_data.groupby(eth_data.index.year):
    group.to_csv(f'eth_data_{year}.csv')
```
